
-   Explain why their solution works in your own words.


```python
# bt_path walks the tree's branches using dfs (depth-first search).   
# dfs function follows the path recursively, first to the left and then to the right.  
# When a leaf is discovered, the path is documented.  Backtracking guarantees that all potential paths are explored.
```

```python
# Your answer here
# the solution is fine, but there is an unnecessary path copying: Only copy at leaf nodes.
# Should add path.pop() after recursive calls ensures that the path list is correctly maintained

# My Suggesntion code
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

    def bt_path(root):
        if not root: return []
        result = []

        def dfs(node, path):
            if not node: return
            path.append(node.val)
            if not node.left and not node.right:
                result.append(path[:])
            dfs(node.left, path)
            dfs(node.right, path)
            path.pop() 

        dfs(root, [])
        return result
```

```python
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)
root.left.left.left = TreeNode(8)  
root.left.left.right = TreeNode(9)   
root.right.left.left = TreeNode(10)  
root.right.right.right = TreeNode(11)  

paths = TreeNode.bt_path(root)
print(f"Paths tree: {paths}")
```
